```python
# importing libraries
import numpy as np

# function to calculate entropy
def sentropy(a):
    value,counts = np.unique(a, return_counts=True)
    norm_counts = counts / counts.sum()
    base = 2 
    entropy_value = -(norm_counts * np.log(norm_counts)/np.log(base)).sum()
    return entropy_value

print("Enter number of numbers you want to have in the array")
num_of_ele = np.int(input())
print("\nEnter the numbers into the array")
arr = []
for i in range(0,num_of_ele):
    arr.append(input())
X = np.array(arr)
print("\nThe entropy is", sentropy(X))
```

## **2.1**

```python
import pandas as pd

df_fm = pd.read_csv('fm.csv')
df_bystate = df_fm.groupby('State')['FMID'].count().sort_values(ascending=False).to_frame().reset_index()
df_bystate.rename(columns={'FMID': '#Farmers Market'}, inplace=True)
df_bystate
```

## **2.2**

```python
df_fm_topfive = df_fm[df_fm['State']=='Massachusetts'].groupby('city')['FMID'].count().sort_values(ascending=False).head().to_frame().reset_index() # by deafult head takes first 5
df_fm_topfive.rename(columns={'city': 'City','FMID': '#Farmers Market'}, inplace=True)
df_fm_topfive
```

## **2.3**

```python
df_fm_subset = df_fm[['State','Credit','WIC','WICcash','SNAP']]
df_fm_subset = df_fm_subset.replace(to_replace = ['Y','N','NaN'],value = [1,0,0])
df_payment = df_fm_subset.melt(id_vars='State')
df_payment.groupby(['State','variable']).sum()
df_payment = df_payment.pivot_table(index=['State','variable'], values='value',aggfunc='sum').reset_index()
df_payment = df_payment.rename(columns={'State':'States','variable':'Payment System','value':'#Farmers Market'})
df_payment
```

## **2.4**

```python
df_division  = pd.read_csv('division.csv')
df_merge = pd.merge(df_fm, df_division, on='State', how='inner')
df_merge = df_merge['Division'].value_counts().to_frame().reset_index()
df_merge.rename(columns={'index': 'Division','Division':'#Framers Markets'}, inplace=True)
df_merge
```
